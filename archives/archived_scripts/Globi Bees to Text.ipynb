First we need to convert the tsv file to csv

```python
import csv

# Open the TSV and CSV files properly handling encoding
with open("/content/drive/MyDrive/DS Capstone/globi-bees-filtered_file.tsv", "r", newline='', encoding="utf-8") as tsv_file, \
     open("/content/drive/MyDrive/DS Capstone/globi_bees - Sheet1.csv", "w", newline='', encoding="utf-8") as csv_file:

    # Read the TSV file using csv.reader with tab delimiter
    tsv_reader = csv.reader(tsv_file, delimiter="\t")

    # Write to the new CSV file using csv.writer
    csv_writer = csv.writer(csv_file, delimiter=",", quotechar='"', quoting=csv.QUOTE_MINIMAL)

    for row in tsv_reader:
        csv_writer.writerow(row)

print("Successfully converted TSV to CSV")
```

```python
def basisofrecordname(value):
  if value == 'PreservedSpecimen' or value == 'preservedspecimen' or value == 'preservedSpecimen' or value == 'PRESERVED_SPECIMEN':
    return 'preserved specimen'
  elif value == 'HumanObservation':
    return 'human observation'
  elif value == 'LabelObservation':
    return 'label observation'
  elif value == 'disected_gut_contents':
    return 'dissected gut contents'
  elif value == 'regurgitated_from_specimen':
    return 'regurgitated from specimen'
  elif value == 'MATERIAL_SAMPLE' or value == 'MaterialSample':
    return 'material sample'
  elif value == 'direct_observation':
    return 'direct observation'
  elif value == 'gut_contents_of_roadkill':
    return 'gut contents of roadkill'
  elif value == 'observed_in_scat':
    return 'observed in scat'
  elif value == 'RefSeq':
    return 'Reference Sequence (RefSeq)'
  elif value == 'FossilSpecimen':
    return 'fossil specimen'
  elif value == 'Exsicatti':
    return 'exsiccati'
  elif value == 'regurgitated_from_specimen/dissected_gut_contents':
    return 'regurgitated from specimen/dissected gut contents'
  else:
    return value
```

```python
def interactiontype(value):
  if value == 'visitsFlowersOf':
    return 'visits the flowers of'
  elif value == 'interactsWith':
    return 'interacts with'
  elif value == 'hasHost':
    return 'is hosted by'
  elif value == 'adjacentTo':
    return 'is adjacent to'
  elif value == 'providesNutrientsFor':
    return 'provides nutrients for'
  elif value == 'preysOn':
    return 'preys on'
  elif value == 'hasDispersalVector':
    return 'has a dispersal vector, which is'
  elif value == 'mutualistOf':
    return 'is a mutualist of'
  elif value == 'hasVector':
    return 'has a vector, which is'
  elif value == 'kleptoparasiteOf':
    return 'is a kleptoparasite of'
  elif value == 'parasiteOf':
    return 'is a parasite of'
  elif value == 'allelopathOf':
    return 'is an allelopath of'
  elif value == 'pathogenOf':
    return 'is a pathogen of'
  elif value == 'createsHabitatFor':
    return 'creates a habitat for'
  elif value == 'endoparasiteOf':
    return 'is an endoparasite of'
  elif value == 'coOccursWith':
    return 'co-occurs with'
  elif value == 'commensalistOf':
    return 'is a commensalist of'
  elif value == 'ectoparasiteOf':
    return 'is an ectoparasite of'
  elif value == 'parasitoidOf':
    return 'is a parasitoid of'
  elif value == 'symbiontOf':
    return 'is a symbiont of'
  elif value == 'ecologicallyRelatedTo':
    return 'is ecologically related to'
  elif value == 'endoparasitoidOf':
    return 'is an endoparasitoid of'
  elif value == 'guestOf':
    return 'is a guest of'
  elif value == 'ectoParasitoid':
    return 'is an ectoparasitoid of'
  elif value == 'livesInsideOf':
    return 'lives inside of'
  else:
    return value
```

```python
def getstring(current, current_null):
  text = ""
  text += "This sample is classified as: "

  if current_null['sourceTaxonKingdomName'] == False:
    text = text + "Kingdom: " + str(current['sourceTaxonKingdomName']) + ", "

  elif current_null['sourceTaxonKingdomId'] == False:
    text = text + "Kingdom id: " + str(current['sourceTaxonKingdomId']) + ", "

  if current_null['sourceTaxonPhylumName'] == False:
    text = text + "phylum: " + str(current['sourceTaxonPhylumName']) + ", "

  elif current_null['sourceTaxonPhylumId'] == False:
    text = text + "phylum id: " + str(current['sourceTaxonPhylumId']) + ", "

  if current_null['sourceTaxonClassName'] == False:
    text = text + "class: " + str(current['sourceTaxonClassName']) + ", "

  elif current_null['sourceTaxonClassId'] == False:
    text = text + "class id: " + str(current['sourceTaxonClassId']) + ", "

  if current_null['sourceTaxonOrderName'] == False:
    text = text + "order: " + str(current['sourceTaxonOrderName']) + ", "

  elif current_null['sourceTaxonOrderId'] == False:
    text = text + "order id: " + str(current['sourceTaxonOrderId']) + ", "

  if current_null['sourceTaxonFamilyName'] == False:
    text = text + "family: " + str(current['sourceTaxonFamilyName']) + ", "

  elif current_null['sourceTaxonFamilyId'] == False:
    text = text + "family id: " + str(current['sourceTaxonFamilyId']) + ", "

  if current_null['sourceTaxonGenusName'] == False:
    text = text + "genus: " + str(current['sourceTaxonGenusName']) + ", "

  elif current_null['sourceTaxonGenusId'] == False:
    text = text + "genus id: " + str(current['sourceTaxonGenusId']) + ", "

  if current_null['sourceTaxonSubgenusName'] == False:
    text = text + "subgenus: " + str(current['sourceTaxonSubgenusName']) + ", "

  elif current_null['sourceTaxonSubgenusId'] == False:
    text = text + "subgenus id: " + str(current['sourceTaxonSubgenusId']) + ", "

  if current_null['sourceTaxonSpeciesName'] == False:
    text = text + "species: " + str(current['sourceTaxonSpeciesName']) + ". "

  elif current_null['sourceTaxonSpeciesId'] == False:
    text = text + "species id: " + str(current['sourceTaxonSpeciesId']) + ". "

  else:
    text += ". "


  if current_null['sourceLifeStageName'] == False:
    text = text + "The organism was in the following life stage: " + str(current['sourceLifeStageName']) + ". "

  elif current_null['sourceLifeStageId'] == False:
    text = text + "The organism's life stage ID is: " + str(current['sourceLifeStageId']) + ". "

  if current_null['sourceSexName'] == False:
    text = text + "The organism's sex is " + str(current['sourceSexName']) + ". "

  if current_null['sourceTaxonRank'] == False and current_null['sourceTaxonName'] == False:
    text = text + "This organism was most specifically classified as having a " + str(current['sourceTaxonRank']) + " of " + str(current['sourceTaxonName']) + ". "

  elif current_null['sourceTaxonName'] == False:
    text = text + "This organism was most specifically classified as a " + str(current['sourceTaxonName']) + ". "


  if current_null['sourceBasisOfRecordName'] == False:
    text = text + "This sample's basis of record is " + basisofrecordname(str(current['sourceBasisOfRecordName'])) + ". "



  text += "This sample " + interactiontype(str(current['interactionTypeName'])) + " the following organism: "



  if current_null['targetTaxonKingdomName'] == False:
    text = text + "Kingdom: " + str(current['targetTaxonKingdomName']) + ", "

  elif current_null['targetTaxonKingdomId'] == False:
    text = text + "Kingdom id: " + str(current['targetTaxonKingdomId']) + ", "

  if current_null['targetTaxonPhylumName'] == False:
    text = text + "phylum: " + str(current['targetTaxonPhylumName']) + ", "

  elif current_null['targetTaxonPhylumId'] == False:
    text = text + "phylum id: " + str(current['targetTaxonPhylumId']) + ", "

  if current_null['targetTaxonClassName'] == False:
    text = text + "class: " + str(current['targetTaxonClassName']) + ", "

  elif current_null['targetTaxonClassId'] == False:
    text = text + "class id: " + str(current['targetTaxonClassId']) + ", "

  if current_null['targetTaxonOrderName'] == False:
    text = text + "order: " + str(current['targetTaxonOrderName']) + ", "

  elif current_null['targetTaxonOrderId'] == False:
    text = text + "order id: " + str(current['targetTaxonOrderId']) + ", "

  if current_null['targetTaxonFamilyName'] == False:
    text = text + "family: " + str(current['targetTaxonFamilyName']) + ", "

  elif current_null['targetTaxonFamilyId'] == False:
    text = text + "family id: " + str(current['targetTaxonFamilyId']) + ", "

  if current_null['targetTaxonGenusName'] == False:
    text = text + "genus: " + str(current['targetTaxonGenusName']) + ", "

  elif current_null['targetTaxonGenusId'] == False:
    text = text + "genus id: " + str(current['targetTaxonGenusId']) + ", "

  if current_null['targetTaxonSubgenusName'] == False:
    text = text + "subgenus: " + str(current['targetTaxonSubgenusName']) + ", "

  elif current_null['targetTaxonSubgenusId'] == False:
    text = text + "subgenus id: " + str(current['targetTaxonSubgenusId']) + ", "

  if current_null['targetTaxonSpeciesName'] == False:
    text = text + "species: " + str(current['targetTaxonSpeciesName']) + ". "

  elif current_null['targetTaxonSpeciesId'] == False:
    text = text + "species id: " + str(current['targetTaxonSpeciesId']) + ". "

  else:
    text += ". "


  if current_null['targetLifeStageName'] == False:
    text = text + "The organism was in the following life stage: " + str(current['targetLifeStageName']) + ". "

  elif current_null['targetLifeStageId'] == False:
    text = text + "The organism's life stage ID is: " + str(current['targetLifeStageId']) + ". "

  if current_null['targetSexName'] == False:
    text = text + "The organism's sex is " + str(current['targetSexName']) + ". "

  if current_null['targetTaxonRank'] == False and current_null['targetTaxonName'] == False:
    text = text + "This organism was most specifically classified as having a " + str(current['targetTaxonRank']) + " of " + str(current['targetTaxonName']) + ". "

  elif current_null['targetTaxonName'] == False:
    text = text + "This organism was most specifically classified as a " + str(current['targetTaxonName']) + ". "

  if current_null['targetBasisOfRecordName'] == False:
    text = text + "This organism's basis of record is " + basisofrecordname(str(current['targetBasisOfRecordName'])) + ". "



  if current_null['decimalLatitude'] or current_null['localityName'] or current_null['localityId'] or current_null['eventDate'] == False:
    text += "The sample was found "

  if current_null['decimalLatitude'] == False:
    text = text + "at latitude: " + str(current['decimalLatitude'].round(2)) + ", longitude: " + str(current['decimalLongitude'].round(2)) + ", "

  if current_null['localityName'] == False:
    text = text + "in " + str(current['localityName']) + ", "

  elif current_null['localityId'] == False:
    text = text + "in the locality id: " + str(current['localityId']) + ", "

  if current_null['eventDate'] == False:
    text = text + "on the date: " + str(current['eventDate'])

  text += ". "

  if current_null['referenceCitation'] == False:
    text = text + "The reference citation is: " + str(current['referenceCitation']) + ". "

  if current_null['referenceDoi'] == False:
    text = text + "The reference DOI is: " + str(current['referenceDoi']) + ". "

  text = text + "The source citation is: " + str(current['sourceCitation']) + ". "

  if current_null['sourceDOI'] == False:
    text = text + "The source DOI is: " + str(current['sourceDOI']) + ". "

  text = text + "The source namespace is: " + str(current['sourceNamespace']) + ". The source URI is: " + str(current['sourceArchiveURI']) + ". The sample was last seen at the unix epoch: " + str(current['sourceLastSeenAtUnixEpoch']) + "."


  text = text + "\n===\n"


  return text
```
